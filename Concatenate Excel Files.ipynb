```python
import pandas as pd
```

```python
# Define Filepath
excel_filepath = r"C:\Users\16476\Desktop\Itamar's Folder\BCIA Class Notes\RSC885 Capstone Project\20230302_OntarioMSBs_IZ.xlsx"
```

```python
# Load Excel File Using Pandas
df1 = pd.ExcelFile(excel_filepath)
```

```python
# Define an Empty List to Store Data Frames Representing Each Worksheet in the Excel File
list_of_dfs = []
```

```python
# Iterate Through Each Worksheet
for worksheet in df1.sheet_names:
    # Parse Data From Each Worksheet and Store in a Pandas Data Frame
    df2 = df1.parse(worksheet)
    # Append the Parsed Data Into the List to Store it in the Pandas Data Frame
    list_of_dfs.append(df2)
```

```python
# Combine Each of the Parsed Data Frames Into a Single Pandas Data Frame
all_jurisdictions = pd.concat(list_of_dfs, ignore_index = True)
```
